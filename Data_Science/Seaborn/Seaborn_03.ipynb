```python
import numpy as np
import pandas as pd
import os

# import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothicBold')
plt.rc('axes', unicode_minus = False)
# plt.rcParams['axes.unicode_minus'] = False

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook")
# sns.reset_defaults()
sns.set_palette(sns.color_palette("pastel"))
```

## seaborn 사용하기

### sns.boxplot: Draw a box plot to show distributions with respect to categories.

```python
titanic = sns.load_dataset("titanic")
```

```python
## Draw a single horizontal boxplot, Group by a categorical variable, referencing columns in a dataframe

fig, axs = plt.subplots(1, 1, figsize = (8, 3))

# sns.boxplot(x=titanic["age"], ax = axs[0])
sns.boxplot(data=titanic, x="age", y="class")

fig.tight_layout()
plt.show()
```

```python
## Draw a vertical boxplot with nested grouping by two variables

fig, axs = plt.subplots(figsize = (6, 4))

sns.boxplot(data=titanic, x="class", y="age", hue="alive", width = 0.6, palette="Paired")

plt.show()
```

```python
## Draw the boxes as line art and add a small gap between them
# kwargs = {
#     'boxprops':{'facecolor':'none', 'edgecolor':'k',
#                 'alpha': 0.5},
#     'medianprops':{'color':'green'},
#     'whiskerprops':{'color':'blue'},
#     'capprops':{'color':'magenta'}
# }

fig, axs = plt.subplots(figsize = (6, 4))

sns.boxplot(data=titanic, x="class", y="age", hue="alive", showfliers=False)
plt.legend(frameon=False)

plt.show()
```

```python
## Cover the full range of the data with the whiskers

fig, axs = plt.subplots(figsize = (6, 4))

sns.boxplot(data=titanic, x="age", y="deck", whis=(0, 100), color="0.8")

plt.show()
```

```python
## Customize the plot using parameters of the underlying matplotlib function

sns.boxplot(
    data=titanic, x="age", y="class",
    notch=True, showcaps=True,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.3, .5, .7, .5)}, # rgb + alpha
    medianprops={"color": "r", "linewidth": 2},
    capprops = {'color':'magenta'}
)
plt.show()
```

### sns.violinplot: Draw a patch representing a KDE (Kernel density estimation) and add observations or box plot statistics

```python
titanic = sns.load_dataset("titanic")

sns.violinplot(x="age", data = titanic)

plt.show()
```

```python
## Violinplot by group
fig, axs = plt.subplots(figsize = (6, 4))

sns.violinplot(data=titanic, x="class", y="age", hue="alive", inner="point",
               palette= "Paired", alpha = 0.5)

plt.show()
```
