####  **결측치 처리하기**

앞에서 배웠듯이 결측치를 처리하는 방법은 크게 세 가지가 있습니다 .

1. 행을 **제거한다 (drop)**
2. 열을 **제거한다 (drop)**
3. **다른 값으로 대체한다 (fill)**


```python
import pandas as pd
import numpy as np

df = pd.DataFrame([ [1,2,np.nan], [3,np.nan,4], [np.nan, 5, 6], [np.nan, np.nan, np.nan]], columns=['a','b','c'])
df
```

:> ❎ **제거(Drop) 하는 법** : 비어있는 값(NaN)이 너무 많을 경우, 전처리가 불가능한 경우 사용
> - `데이터프레임.dropna()`
> - `데이터프레임.dropna(how='all')`

```python
df.dropna(axis=0, how='any')
```

```python
df.dropna(axis=0, how='all')
```

```python
df.dropna(axis=1,how='any')
```

```python
df.dropna(axis=0, how='all', inplace=True)
```

```python
df
```

```python
df1 = df.dropna(axis=0, how='all')
```

```python
df1
```

##### 🎨 **결측치 대체하기**

결측치에 평균값을 임의로 넣어주는 평균대체법💁🏼, 결측치에 회귀분석을 이용한 예측값을 넣어주는 회귀대체법🎯 등이 있답니다! 

```python
titanic['age']
```

>통계적으로 차이가 나지 않도록 **평균(mean)** 에 **분산(std)** 만큼 차이나는 사이의 임의값을 넣어주기!

Age의 경우 `.drop()`을 사용해서 버리기에는 너무 중요한 값입니다.   

그렇다고 일괄적으로 빈칸에 0 등의 값을 넣으면 오히려 자료가 왜곡되는 문제가 생깁니다.

```python
sns.histplot(titanic['age'])
plt.show()
```

```python
# 결측치를 0으로 바꾼 뒤 age의 분포
sns.histplot
plt.show()
```


그래서 이런 숫자 데이터에 일반적으로 **자료의 분포를 해치지 않는 값**을 생성하여 집어넣습니다

대표적으로, 자료와 평균과 분산이 일치하는 **정규분포**에서 표본을 생성하여 집어넣는 방법이 있죠 

<img src='https://i.postimg.cc/1tWLZvzP/image.png'  width="400px;" border='0'>

 <----- 데이터 ----->  
- 평균 = 데이터의 중앙에 있을 확률이 높은 값  
- 분산 = 각 값이 평균에서 얼마만큼 분산되어있는지의 척도   

그래서 평균에서 분산만큼 떨어진 값들 사이가 대략적으로 가장 확률이 높은 곳이라고 볼 수 있습니다.  

`분산  ~ 평균() ~ 분산 `

이러한 방법은 실제 자료가 정규분포를 따를 것이라 가정하고, 관측하지 못한 자료가 위치할 것으로 예상되는 곳에서 그 자료를 복원해내는 방법이라 할 수 있습니다.
