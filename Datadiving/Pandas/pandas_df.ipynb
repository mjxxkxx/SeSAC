 > 문제1

 - Survived 가 0
 - Pclass가 3등급
 - Sex가 female  
인 데이터만 필터링!!

```python
titanic_df[(titanic_df['Survived']==0)&
           (titanic_df['Pclass']==3) &
           (titanic_df['Sex']=='female')
]
```

`len()`을 활용하면 필터링된 데이터의 총량을 알 수 있다.
> Pclass별 여성의 생존자 수

```python
print('Female survivor (1st class): ', len(titanic_df[(titanic_df['Pclass']==1) & 
(titanic_df['Survived']==1)& (titanic_df['Sex']=='female')]))

print('Female survivor (2nd class): ', len(titanic_df[(titanic_df['Pclass']==2) & (titanic_df['Survived']==1)& (titanic_df['Sex']=='female')]))

print('Female survivor (3rd class): ',len(titanic_df[(titanic_df['Pclass']==3) & (titanic_df['Survived']==1)& (titanic_df['Sex']=='female')]))
```

```python
# use Pclass as index label
titanic_groupby = titanic_df.groupby('Pclass').count()
titanic_groupby
```
