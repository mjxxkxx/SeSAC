```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial(x, n):
    y = x**n
    return y

x = np.linspace(-4, 4, 100)
# print(x)
y = polynomial(x, 1) # linear fun
y2 = polynomial(x, 2)
y3 = polynomial(x, 3)
y4 = polynomial(x, 4)

plt.plot(x, y, label = "y = x")
plt.plot(x, y2, label = "y = x^2")
plt.plot(x, y3, label = "y = x^3")
plt.plot(x, y4, label = "y = x^4")

plt.title("Polynomial Curve")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid(linestyle= ":")
plt.show()
```

## 삼각함수

```python
print(np.pi)
```

```python
X = np.linspace(-2*np.pi, 2*np.pi, 100)
# print(X)
sin_x = np.sin(X)
cos_x = np.cos(X)
tan_x = np.tan(X)
plt.figure(figsize = (12, 5))
plt.plot(X, sin_x, label = "sin(x)")
plt.plot(X, cos_x, label = "cos(x)")
plt.plot(X, tan_x, label = "tan(x)")
plt.vlines(0, -1, 1, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")

plt.ylim(-1, 1)
plt.xlim(-2*np.pi, 2*np.pi)
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])

plt.legend()
plt.grid(linestyle = ":")
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)

sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

y_list = [sin_x, cos_x, tan_x]
y_label = ["sin(x)", "cos(x)", "tan(x)"]
color_list = ["b", "g", "r"]
fig, ax = plt.subplots(3, 1, figsize = (12, 10)) # 3행 1열


for i in range(3):
    ax[i].plot(x, y_list[i], label = y_label[i], c = color_list[i])
    ax[i].set_ylim(-1, 1)
    ax[i].grid(linestyle = ":")
    ax[i].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
    ax[i].vlines(0, -1, 1, color = "r")
    ax[i].hlines(0, -2*np.pi, 2*np.pi, color = "k")

plt.show()
```

```python
## (1) y - b = sin(x - a) 의 그래프는 y = sin(x) 의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.
## (1) y = sin(x - a) + b 의 그래프는 y = sin(x) 의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = np.sin(x - 2* np.pi) + 1 # x축으로 +pi, y축으로 +1 평행이동

plt.figure(figsize = (12, 5))
plt.plot(x, y, label = "sin(x)", color = "k")
plt.plot(x, y1, label = "sin(x - 2pi) + 1")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## y = -sin(x) 의 그래프는 y = sin(x) 의 그래프를 x축으로 선대칭이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = -np.sin(x) 

plt.figure(figsize = (12, 5))
plt.plot(x, y, label = "sin(x)", color = "k")
plt.plot(x, y1, label = "-sin(x)", linestyle = ":")
plt.vlines(0, -1, 1, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## y = -cos(x) 의 그래프는 y = cos(x) 의 그래프를 y축으로 선대칭이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = -np.cos(x) 

plt.figure(figsize = (12, 5))
plt.plot(x, y, label = "sin(x)", color = "k")
plt.plot(x, y1, label = "-sin(x)", linestyle = ":")
plt.vlines(0, -2, 2, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "r")

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## (3) y = acos(x)의 그래프는 y = cos(x) 의 함숫값을 a배 한 것이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = 2*np.cos(x) 

plt.figure(figsize = (12, 5))
plt.plot(x, y, label = "cos(x)", color = "k")
plt.plot(x, y1, label = "2cos(x)", linestyle = ":")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "r")

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## (4) y = cos(ax)의 그래프는 y = cos(x)와 모양은 같고 주기가 2pi/|a| 인 그래프이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = np.cos(2*x) 

plt.figure(figsize = (12, 5))
plt.plot(x, y, label = "cos(x)", color = "k")
plt.plot(x, y1, label = "cos(2*x)", linestyle = ":")
plt.vlines(0, -1, 1, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "r")

plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -0.5*np.pi, 0, 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["-2pi", "-1.5pi", "-pi", "-0.5pi", "0", "0.5pi", "pi", "1.5pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
##  y = sin(x - pi/2) 와 y = sin(x - pi/2) + 1의 그래프를 그려라.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.sin(x - 0.5 *np.pi) 
y2 = np.sin(x - 0.5 *np.pi) + 1

plt.figure(figsize = (12, 5))
plt.plot(x, y1, label = "sin(x - pi/2)", color = "k")
plt.plot(x, y2, label = "sin(x - pi/2) + 1")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")


plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -0.5*np.pi, 0, 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["-2pi", "-1.5pi", "-pi", "-0.5pi", "0", "0.5pi", "pi", "1.5pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## y = 2sinx 와 y = 2sin(x - pi/2) + 1

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = 2 * np.sin(x) 
y2 = 2 * np.sin(x - 0.5 *np.pi) + 1

plt.figure(figsize = (12, 5))
plt.plot(x, y1, label = "2sin(x)", color = "k")
plt.plot(x, y2, label = "2sin(x - pi/2) + 1")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")


plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -0.5*np.pi, 0, 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["-2pi", "-1.5pi", "-pi", "-0.5pi", "0", "0.5pi", "pi", "1.5pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## y = sin2x 와 y = 2sin(x - pi/2) - 1

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.sin(2*x) 
y2 = 2 * np.sin(x - 0.5 *np.pi) - 1

plt.figure(figsize = (12, 5))
plt.plot(x, y1, label = "sin(2x)", color = "k")
plt.plot(x, y2, label = "2sin(x - pi/2) - 1")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")


plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -0.5*np.pi, 0, 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["-2pi", "-1.5pi", "-pi", "-0.5pi", "0", "0.5pi", "pi", "1.5pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## y = cosx 의 그래프에 대하여 어떤 변화인지 말해라~

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.cos(x - np.pi/2) 
y2 = -np.cos(x) + 1 
y3 = np.cos(2*x + np.pi) 

plt.figure(figsize = (12, 5))
plt.plot(x, y1, label = "cos(x - pi/2)")
plt.plot(x, y2, label = "-cos(x) - 1")
plt.plot(x, y3, label = "cos(2x + pi)")
plt.vlines(0, -3, 3, color = "r")
plt.hlines(0, -2*np.pi, 2*np.pi, color = "k")


plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -0.5*np.pi, 0, 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["-2pi", "-1.5pi", "-pi", "-0.5pi", "0", "0.5pi", "pi", "1.5pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```
