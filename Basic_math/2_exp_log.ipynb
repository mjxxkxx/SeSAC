```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 100)
y = 2**x
y1 = 2**(x-1)
y2 = 2**(x-1) + 1
y3 = 2**(-(x-2))
y4 = 2**(-x)

plt.plot(x, y, label = "2^x")
plt.plot(x, y1, label = "2^(x-1)")
plt.plot(x, y2, label = "2^(x-1)+ 1", linestyle = ":")
plt.plot(x, y3, label = "2^(-(x-2))", linestyle = ":")
plt.plot(x, y4, label = "2^(-x)", linestyle = ":")

plt.hlines(0, -5, 5, color = "k")
plt.vlines(0, -1, 70, color = "r")
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = 2 ** (x)
y_exp = np.exp(x)
y3 = 3 ** (x)
y_exp2 = np.exp(-x)


plt.plot(x, y, label = "2^x", linestyle = ":")
plt.plot(x, y_exp, label = "e^x")
plt.plot(x, y_exp2, label = "e^(-x)")
plt.plot(x, y3, label = "3^x", linestyle = ":")

plt.hlines(0, -5, 5, color = "k")
plt.vlines(0, -1, 25, color = "r")
plt.grid(linestyle = ":")
plt.legend()
plt.show()
```

```python
## Binary loss function
x = np.linspace(0, 4, 100)
y1 = -np.log(x)
y2 = -np.log(1 - x)

plt.plot(x, y1, label = "-log(x)")
plt.plot(x, y2, label = "-log(1-x)")

plt.grid(linestyle = ":")
plt.xlim(0, 1)
plt.vlines(1, -2, 3, linestyles= ":")
plt.vlines(0, -2, 3, color = "k")
plt.hlines(0, 0, 4, color = "k")
plt.legend()
plt.show()
```
